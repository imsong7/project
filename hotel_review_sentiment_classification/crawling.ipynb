```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd
import csv
```

```python
driver=webdriver.Chrome("C:/Temp/chromedriver.exe")
```

```python
#아리랑힐 호텔 동대문
need_reviews_cnt = 50
reviews1 = []
review_data1=[]
review_score1=[]

for i in range(0,10):
    url = f'https://www.tripadvisor.co.kr/Hotel_Review-g294197-d8472412-Reviews-or{i*5}-Arirang_Hill_Hotel_Dongdaemun-Seoul.html#REVIEWS'
    #get : request로 url의  html문서의 내용 요청
    driver.get(url)
    #html을 받아온 문서를 .content로 지정 후 soup객체로 변환
    html=driver.page_source
    soup=BeautifulSoup(html,'html.parser')

    #평점 가져오기
    reviews1_score = soup.find_all("div",{"class":"Hlmiy F1"})
    for review_s in reviews1_score:
        reviews1_score = str(review_s).split('ui_bubble_rating bubble_')
        reviews1_score = reviews1_score[1]

        review_score1.append(reviews1_score[:1])

    time.sleep(0.5)

    #find_all : 지정한 태그의 내용을 모두 찾아 리스트로 반환
    reviews1 = soup.find_all("span",{"class":"QewHA H4 _a"})
    for review in reviews1:
        sentence = review.get_text()
        if sentence != "":
            review_data1.append(sentence)
            need_reviews_cnt-= 1
            #현재까지 수집된 리뷰가 목표 수집 리뷰보다 많아진 경우 크롤링 중지
    if need_reviews_cnt < 0:
        break
    #다음 페이지를 조회하기 전 0.5초 시간 차를 두기
    time.sleep(0.5)

review1 = pd.DataFrame(zip(review_score1, review_data1))
review1['hotel']="아리랑힐 호텔 동대문"
review1.columns = ['score', 'review', 'hotel']
review1=review1[['hotel', 'score', 'review']]
review1
```
